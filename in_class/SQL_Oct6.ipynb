### Join
A list of every album title paired with the artist name that made it. One row per album that has a matching artist.

```python


query_4 = '''
SELECT Title, Name
FROM albums a
INNER JOIN artists r
ON (r.ArtistId = a.ArtistId)
'''
query_4 = pd.read_sql_query(query_4, chinook)
query_4.head()

```

# Chains joins and filter
“Show me up to 100 songs starting with ‘D,’ with their album and artist info.”

```python
query_5 = '''
SELECT t.Name as Track,
        Milliseconds as TrackMS,
        Title as Album,
        r.Name as Artist
FROM tracks t
INNER JOIN albums a
ON (t.AlbumId = a.AlbumId)
INNER JOIN artists r
ON (r.ArtistId = a.ArtistId)
WHERE Track LIKE "D%"
LIMIT 100
'''
query_5 = pd.read_sql_query(query_5, chinook)
query_5.head()
```
