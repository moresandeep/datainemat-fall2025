## 1. Load & Peek

## 2. Select Columns
Quick ways to select columns: `[]`, `.loc`, `.iloc`, `.filter()`, `select_dtypes()`

```python
# Single & multiple
tips['total_bill'].head()
```

```python
tips[['total_bill','tip']].head()
```

```python
# Label vs position
tips.loc[0:5, ['day','time','total_bill']]
```

```python
tips.iloc[0:5, 0:3]  # first 5 rows & first 3 columns
```

```python
# By dtype
tips.select_dtypes(include='number').head()
```

```python
tips.select_dtypes(include=['object','category']).head()
```

```python
# Pattern match
tips.filter(like='ti').head()
```

## 3. Filter Rows (Boolean Operators)
Use `&` and `|` with parentheses.

```python
# Numeric comparisons & logical AND
tips[(tips['total_bill'] > 20) & (tips['tip'] >= 3)].head()
```

```python
# Membership (OR within a set)
tips[tips['day'].isin(['Sat','Sun'])].head()
```

```python
# NOT (~) and null checks
tips[~tips['sex'].isna()].head()
```

## 4. `DataFrame.query()` — readable, SQL-ish filtering

```python
target_days = ['Sat','Sun']
tips.query("total_bill > 20 and day in @target_days").head()
```

## 5. Method Chaining → Quick Bar Chart

```python
(
    tips
    .assign(tip_pct=lambda d: d['tip'] / d['total_bill'] * 100)
    .groupby('day', dropna=False)['tip_pct'].mean()
    .reset_index()
    .sort_values('tip_pct', ascending=False)
)
```

```python
(
    tips['day']
    .value_counts()
    .reset_index(name='count')
    .rename(columns={'index':'day'})
    .plot(kind='bar', x='day', y='count', legend=False, title='Rows per Day')
); plt.show()
```

## 6. Common Pitfalls
- Use `&`/`|`, not `and`/`or`
- Always parenthesize combined conditions
- Prefer `.loc`/`.iloc` over chained indexing
- Null‑aware string filters: `na=False`
- Backticks in `query()` for awkward column names
