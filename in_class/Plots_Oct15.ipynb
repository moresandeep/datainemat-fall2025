# Univariate Visualizations

## Histogram

```python
# Histogram
plt.hist(tips["tip"])
plt.title("Distribution of Tip Amounts")
plt.xlabel("Tip ($)")
plt.ylabel("Frequency")
plt.show()
```

## Boxplot

Where are most tips concentrated? Any unusually high or low tips?

```python
sns.boxplot(x=tips["tip"])
plt.title("Boxplot of Tips")
plt.show()
#plt.figure(figsize=(6,2))
#plt.boxplot(tips["tip"], vert=False, patch_artist=True)
#plt.title("Boxplot of Tips")
#plt.xlabel("Tip ($)")
#plt.show()
```

Where is the median tip? Are there any outliers?
	•	Most customers tipped between $2 and $4.
	•	The median tip is about $3.
	•	A few customers gave very high tips (>$6), shown as outliers.
	•	The distribution is right-skewed — most tips are small, with a few large tips stretching the upper tail.

The box spans from about $2 (Q1) to $4 (Q3).
This means that 50% of all tips fall between $2 and $4.
The orange line inside the box is around $3, meaning half the tips are below $3 and half are above.

## Violin Plot

```python
sns.violinplot(x=tips["tip"])
plt.title("Violin Plot of Tips")
plt.show()
```

## Density Plot

```python
sns.kdeplot(tips["tip"], fill=True)
plt.title("Density Plot of Tips")
plt.show()
```

# Multivariate Visualizations

## Scatter Plot

```python

sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Tip vs Total Bill")
plt.show()
```

As the bill increases, do tips increase too?

## Boxplots by Category

```python
sns.boxplot(x="day", y="tip", data=tips)
plt.title("Tip Distribution by Day")
plt.show()
```

# Mapping and Spatial Thinking 

```python
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.csv"
df = pd.read_csv(url)
```

```python
# Quick look
df.head()[["place", "mag", "time", "latitude", "longitude"]]
```

```python
# Simple scatter map
plt.scatter(df["longitude"], df["latitude"], 
            s=df["mag"]**3,   # bubble size reflects magnitude
            edgecolors="black")
plt.title("Earthquakes in the Last 24 Hours")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
